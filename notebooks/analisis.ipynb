```python
import pandas as pd
import matplotlib.pyplot as plt

import sys
import os

parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir)) # ../
sys.path.append(parent_dir)

# 1. Carga y limpieza bsica
df = pd.read_csv(r'C:\Users\aceve\OneDrive\Documents\local_development\analisis-encuestas-stack-overflow\analisis-encuestas-stack-overflow\data\raw\survey_results_public.csv', low_memory=False)
df = df.drop_duplicates()
df = df.dropna(subset=['CompTotal', 'EdLevel', 'WorkExp'])

# 2. Figura 1: Histograma de compensacin anual (CompTotal)
plt.figure(figsize=(10,6))
plt.hist(df['CompTotal'], bins=30)
plt.title('Distribucin de compensacin anual (CompTotal)')
plt.xlabel('Compensacin anual (USD)')
plt.ylabel('Frecuencia')
plt.tight_layout()
#plt.savefig('/reports/images/compensacion_histograma.png')
#plt.close()

# 3. Figura 2: Mediana de compensacin por nivel educativo (EdLevel)
medianas = df.groupby('EdLevel')['CompTotal'].median().sort_values()
plt.figure(figsize=(10,6))
medianas.plot(kind='bar')
plt.title('Mediana de compensacin por nivel educativo')
plt.xlabel('Nivel educativo')
plt.ylabel('Compensacin mediana (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
#plt.savefig('reports/images/salario_educacion_mediana.png')
#plt.close()

# 4. Figura 3: Histograma de aos de experiencia (WorkExp)
plt.figure(figsize=(10,6))
plt.hist(df['WorkExp'], bins=30)
plt.title('Distribucin de aos de experiencia (WorkExp)')
plt.xlabel('Aos de experiencia')
plt.ylabel('Frecuencia')
plt.tight_layout()
#plt.savefig('reports/images/experiencia_histograma.png')
#plt.close()

print("Figuras generadas y guardadas en reports/images/")
```
